```python
unique_Class_v12_code = np.unique(df['Class_v12_code'])
indices = []

# Create a dictionary to count occurrences
counts = {code: 0 for code in unique_Class_v12_code}

for i, code in enumerate(df['Class_v12_code']):
    if counts[code] < 3:
        indices.append(i)
        counts[code] += 1

```

```python
dfsmall = df.iloc[indices]
```

```python
dfsmall.head(30)
```

```python

validation = dfsmall['Class_v12_code'].value_counts() == 3
assert validation.all(), 'not exactly 3 for each class'
print('test passed all classes have exactly 3') 
dfsmall.sort_values('Class_v12_code').head(30)
```

```python
first_three_columns = dfsmall.iloc[:, :3]
sliced_columns = dfsmall.iloc[:, 3::100]  # Starting from 4th column, take every 500th column

# Concatenate the two sets of columns
dfsmall_2 = pd.concat([first_three_columns, sliced_columns], axis=1)
```
